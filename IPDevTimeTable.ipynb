```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
df_raw=pd.read_csv('./IPDevTimeTable_distinct.csv',header=0)
df_raw=df_raw[df_raw.ip!='ip']#.set_index("ip")#.reset_index()#.drop("index",inplace=True)
df_raw["ip&cnt"]=df_raw['ip']+"_"+df_raw['devcnt']
df_raw.drop(["ip"],axis=1,inplace=True)#.set_index("ip&cnt")
df_raw=df_raw.set_index("ip&cnt")
df_raw["devcnt"]=df_raw["devcnt"].astype(np.int32)
df_raw.head()
```

```python
df_raw[df_raw.devcnt.astype(np.int32)<1000].head(20)
```

```python
df_raw[df_raw["devcnt"].astype(np.int32) > 100000].sum()[1:].plot()
```

```python
df_raw[df_raw["devcnt"].astype(np.int32) < 100000].sum()[1:].plot()
```

```python
fig,ax=plt.subplots(figsize=(20,12))
df_raw[df_raw["devcnt"].astype(np.int32) > 100000].sort_values(by='devcnt',ascending=False).iloc[:,1:].T.plot(ax=ax)
```

```python
fig,ax2=plt.subplots(figsize=(20,12))
df_raw[df_raw["devcnt"].astype(np.int32) < 100000].sort_values(by='devcnt',ascending=False).iloc[:,1:].T.plot(ax=ax2)
```

```python
#df_raw[df_raw["devcnt"].astype(np.int32) < 100000]#.sort_values(by='devcnt',ascending=False).iloc[:,1:]/df_raw['devcnt']
```

```python
df_raw_rate=df_raw.div(df_raw.devcnt, axis=0)
df_raw_rate["devcnt"]=df_raw["devcnt"]
df_raw_rate.head()
```

```python
fig,ax3=plt.subplots(figsize=(20,12))
df_raw_rate[df_raw_rate["devcnt"].astype(np.int32) > 100000].sort_values(by='devcnt',ascending=False).iloc[:,1:].T.plot(ax=ax3)
```

```python
fig,ax4=plt.subplots(figsize=(20,12))
df_raw_rate[df_raw_rate["devcnt"].astype(np.int32) < 100000].sort_values(by='devcnt',ascending=False).iloc[:,1:].T.plot(ax=ax4)
```
