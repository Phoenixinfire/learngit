```python
p='./input'
import os
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import seaborn as sns
```

```python
df_iris=pd.read_csv(os.path.join(p,'Iris.csv'))
df_iris.head()
```

```python
df_iris.Species.value_counts()
```

```python
df_iris.plot(kind="scatter",x='SepalLengthCm',y='SepalWidthCm',grid=True)
```

```python
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=df_iris,kind='scatter')
```

```python
grid=sns.FacetGrid(df_iris,hue="Species",size=6)
grid.map(plt.scatter,"SepalLengthCm",'SepalWidthCm').add_legend()
```

```python
sns.boxplot(x='Species',y='PetalLengthCm',data=df_iris)
```

```python
sns.boxplot(x='Species',y='PetalLengthCm',data=df_iris)
sns.stripplot(x='Species',y='PetalLengthCm',data=df_iris,jitter=True)
```

```python
sns.violinplot(x="Species", y="PetalLengthCm",data=df_iris)
```

```python
grid=sns.FacetGrid(df_iris,hue='Species',size=6)
grid.map(sns.kdeplot,'PetalLengthCm').add_legend()
```

```python
sns.pairplot(data=df_iris.drop('Id',axis=1),hue="Species",size=3)
```

```python
sns.pairplot(data=df_iris.drop('Id',axis=1),hue="Species",size=3,diag_kind='kde')
```

```python
df_iris.drop('Id',axis=1).boxplot(by="Species",figsize=(12,6))
```
