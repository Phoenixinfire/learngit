```python
p_train='./input/train'
p_test='./input/test'
import os
import sys
import cv2
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
import numpy as np
%matplotlib inline
```

```python
?convolve2d
```

```python
image=cv2.imread(os.path.join(p_train,'cat.46.jpg'))
#converting the image to grayscale
image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
image.shape
```

```python
def show_differences(kernel):
    convolved=convolve2d(image,kernel)
    fig=plt.figure(figsize=(15,15))
    plt.subplot(121)
    plt.title("Original image")
    plt.axis('off')
    plt.imshow(image,cmap='gray')
    
    plt.subplot(122)
    plt.title("Convolved image")
    plt.axis('off')
    plt.imshow(convolved,cmap='gray')
    return convolved
```

```python
kernel=np.ones((8,8),np.float32)/64
output=show_differences(kernel=kernel)
```

```python
kernel=np.array([[-1,0,1],[-2,0,-2],[-1,0,1]],np.float32)
dx=show_differences(kernel)
```

```python
kernel=np.array([[1,2,1],[0,0,0],[-1,-2,-1]],np.float32)
dy=show_differences(kernel)
```

```python
mag=np.hypot(dx,dy)
mag*=255.0/np.max(mag)
fig = plt.figure(figsize=(15, 15))
plt.subplot(121)
plt.title('Original image')
plt.axis('off')
plt.imshow(image, cmap='gray')

plt.subplot(122)
plt.title('Convoluted image with highlighted edges')
plt.axis('off')
plt.imshow(mag, cmap='gray')
```
